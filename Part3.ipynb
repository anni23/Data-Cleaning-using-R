Question:What is the number of Males and Females of different age who use dating websites? 

Observation:More number of males in general use dating websites as compared to number of females in all age group.
And in general more people in age group of 20 to 39 use the dating website.

```r
c1=c(
nrow(filter(df,sex==1,age<20,date1a==1)),
nrow(filter(df,sex==1,age>=20,age<40,date1a==1)),
nrow(filter(df,sex==1,age>=40,age<60,date1a==1)),
nrow(filter(df,sex==1,age>=60,age<80,date1a==1)),
nrow(filter(df,sex==1,age>=80,age<100,date1a==1))
)
c1



c2=c(
nrow(filter(df,sex==2,age<20,date1a==1)),
nrow(filter(df,sex==2,age>=20,age<40,date1a==1)),
nrow(filter(df,sex==2,age>=40,age<60,date1a==1)),
nrow(filter(df,sex==2,age>=60,age<80,date1a==1)),
nrow(filter(df,sex==2,age>=80,age<100,date1a==1))
)
c2


Age=c('0-19','20-39','40-59','60-79','80-99')
Num=c(1,2,3,4,5)
b1=cbind(Num,c1)
d1=as.data.frame(b1)
b2=cbind(Num,c2)
d2=as.data.frame(b2)


#ggplot(data=d1,aes(x=Num,y=c1))
plot(d1,type='o',main='Age wise distribution of Males and Females who use dating websites',xaxt='n',col='red',ylab='Number of people',xlab='Age distribution')
lines(d2,col='blue',type='o')
axis(side=1,at=Num,labels=Age)
legend(4, 50 ,legend=c("Males", "Females"),
       col=c("red", "blue"), lty=1:1, cex=0.8)
```

Question:What is the purpose of using internet by people of different age?

Obsevation: In general more people in all age groups use internet for the purpose of searching and applying for jobs.
And more people in age group of 20 to 39 use internet in general.

```r
infojob=c(
nrow(filter(df,age<20,act19==1)),
nrow(filter(df,age>=20,age<40,act19==1)),
nrow(filter(df,age>=40,age<60,act19==1)),
nrow(filter(df,age>=60,age<80,act19==1)),
nrow(filter(df,age>=80,age<100,act19==1))
)
infojob



sn=c(
nrow(filter(df,age<20,act87a==1)),
nrow(filter(df,age>=20,age<40,act87a==1)),
nrow(filter(df,age>=40,age<60,act87a==1)),
nrow(filter(df,age>=60,age<80,act87a==1)),
nrow(filter(df,age>=80,age<100,act87a==1))
)
sn

sm=c(
nrow(filter(df,age<20,act87b==1)),
nrow(filter(df,age>=20,age<40,act87b==1)),
nrow(filter(df,age>=40,age<60,act87b==1)),
nrow(filter(df,age>=60,age<80,act87b==1)),
nrow(filter(df,age>=80,age<100,act87b==1))
)
sm

applyjob=c(
nrow(filter(df,age<20,act137==1)),
nrow(filter(df,age>=20,age<40,act137==1)),
nrow(filter(df,age>=40,age<60,act137==1)),
nrow(filter(df,age>=60,age<80,act137==1)),
nrow(filter(df,age>=80,age<100,act137==1))
)
applyjob



Age=c('0-19','20-39','40-59','60-79','80-99')
Num=c(1,2,3,4,5)
b1=cbind(Num,infojob)
d1=as.data.frame(b1)
b2=cbind(Num,sn)
d2=as.data.frame(b2)
b3=cbind(Num,sm)
d3=as.data.frame(b3)
b4=cbind(Num,applyjob)
d4=as.data.frame(b4)


plot(d1,type='o',main='Age wise distribution of people who use internet for diff purposes',xaxt='n',col='red',ylab='Number of people',xlab='Age distribution')
lines(d2,col='blue',type='o')
lines(d3,col='green',type='o')
lines(d4,col='yellow',type='o')
axis(side=1,at=Num,labels=Age)
legend(3.08, 440 ,legend=c("Look online for information about a job", "Use a social networking site","Use social media","Apply for a job online"),
       col=c("red", "blue","green","yellow"), lty=1:1, cex=0.8)
```
