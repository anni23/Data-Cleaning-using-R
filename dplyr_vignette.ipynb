Work through a vignette for dplyr

```r
library('dplyr')
```

```r
mydb=src_sqlite("database.sqlite")
```

```r
tbl(mydb,sql("select * from Player where height>200"))
```

```r
player_table=tbl(mydb,'Player')
```

```r
select(player_table,id,player_name)
```

```r
filter(player_table,id<10)
```

```r
arrange(player_table,height)
```

```r
c1=select(player_table,weight,height)
mutate(c1,weight,height,test=weight+height)
```

```r
summarise(player_table,avg_height=mean(height))
```

```r
df=collect(player_table)
df$player_name
```
