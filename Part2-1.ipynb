Using src_sqlite() function to connect to the database.
And tbl() function is used for getting reference to the tables in the database.

```r
mydb=src_sqlite("database.sqlite")
player_table=tbl(mydb,'Player')
team_table=tbl(mydb,'Team')
teamAtr_table=tbl(mydb,'Team_Attributes')
playerAtr_table=tbl(mydb,'Player_Attributes')
league_table=tbl(mydb,'league')
match_table=tbl(mydb,'Match')
```

Removing redundant data from the table Player_Attributes by selecting data that has the maximum overall rating.

```r
playerAtr_table1=tbl(mydb,sql("select * from Player_Attributes group by player_api_id having max(overall_rating)"))
```

Joining the tables player_table and playerAtr_table by player_api_id to get player id, player name, overall rating and gk attributes in one table ie test

```r
test=select(merge(player_table,playerAtr_table1,by="player_api_id"),player_name,player_api_id,overall_rating,gk_diving,gk_handling,gk_kicking,gk_positioning,gk_reflexes)
```

Using merge() function to sum up all the gk attributes in test table to get the table test1

```r
test1=mutate(test,gk_sum=gk_diving+gk_handling+gk_kicking+gk_positioning+gk_reflexes)
```

Selecting player id, player name, overall rating and sum of gk attributes from test1 to get the table test2

```r
test2=select(test1,player_name,player_api_id,overall_rating,gk_sum)
```

Here I am using group by and summarise clause to group the table h1 by home_team_api_id and then taking sum over home_team_goals.
This will give me a table containing all team ids along with the number of goals they scored as a home team.

```r
h1=select(match_table,home_team_api_id,home_team_goal)
h2=group_by(h1,home_team_api_id)
#collect(h2)
h3=summarise(h2,sum(home_team_goal))
h4=collect(h3)
colnames(h4)[1]='team_api_id'
colnames(h4)[2]='goals'
h4
nrow(as.data.frame(h1))
nrow(as.data.frame(h4))
```

Here I am using group by and summarise clause to group the table a1 by away_team_api_id and then taking sum over away_team_goals.
This will give me a table containing all team ids along with the number of goals they scored as a away team.

```r
a1=select(match_table,away_team_api_id,away_team_goal)
a2=group_by(a1,away_team_api_id)
#collect(a2)
a3=summarise(a2,sum(away_team_goal))
a4=collect(a3)
colnames(a4)[1]='team_api_id'
colnames(a4)[2]='goals'
a4
nrow(as.data.frame(a1))
nrow(as.data.frame(a4))
```

Concatenating both the dataframes obtained above rows wise.

```r
t1=rbind(h4,a4)
```

Again using group by and summarise clause to group the table t1 obtained above by team_api_id and then taking sum over goals to take into account same team playing sometimes as home team and sometimes as away team.
This will give me a table containing all team ids along with the number of goals they scored.

```r
t2=group_by(t1,team_api_id)
#collect(a2)
t3=summarise(t2,sum(goals))
t4=collect(t3)
colnames(t4)[2]='goals'
t4
nrow(as.data.frame(t1))
nrow(as.data.frame(t4))
```

Getting sum of buildUpPlay attributes, chanceCreation attributes and defence attributes of teams from teamAtr_table

```r
ta=collect(teamAtr_table)
ta[is.na(ta)] <- 0
ta1=mutate(ta,buildUpPlay_sum=buildUpPlaySpeed+buildUpPlayDribbling+buildUpPlayPassing)
ta2=mutate(ta1,chanceCreation_sum=chanceCreationPassing+chanceCreationCrossing+chanceCreationShooting)
ta3=mutate(ta2,defence_sum=defencePressure+defenceAggression+defenceTeamWidth)
collect(ta3)
```

Selecting only the required columns from ta3 into ta4.

```r
ta4=select(ta3,team_api_id,buildUpPlay_sum,chanceCreation_sum,defence_sum)
ta5=collect(ta4)
nrow(ta5)
```

Removing redundant data from ta5.

```r
finalteam=subset(ta5,!duplicated(team_api_id))
```

Merging the dataframes finalteam (containing sum of team attributes) and t4 (containing number of goals scored by each team)

```r
finalteam1=merge(finalteam,t4,by="team_api_id")
```

Merging the dataframe obtained above with the main team_table.

```r
v=collect(team_table)
finalteam2=merge(finalteam1,v,by="team_api_id")
finalteam2
```
